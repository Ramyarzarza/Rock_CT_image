```python
from PIL import Image, ImageDraw
import random

def generate_fracture_image(width=512, height=512, m_num_fractures=50, save_path="fracture_image.png"):
    # Create a grayscale image with a dark gray background
    image = Image.new("L", (width, height), color=50)  # 'L' mode for grayscale, background = 50 (dark gray)
    mask = Image.new("L", (width, height), color=0)
    draw = ImageDraw.Draw(image)
    drawm = ImageDraw.Draw(mask)
    num_fractures = random.randint(0, m_num_fractures)
    # Generate random fractures
    for _ in range(num_fractures):
        # Random starting point for the fracture
        start_x = random.randint(0, width)
        start_y = random.randint(0, height)
        
        # Random length and direction
        length = random.randint(20, 100)
        angle = random.uniform(0, 2 * 3.14159)  # Random angle in radians
        
        # Calculate the end point
        end_x = int(start_x + length * random.uniform(0.5, 1.0) * random.choice([-1, 1]))
        end_y = int(start_y + length * random.uniform(0.5, 1.0) * random.choice([-1, 1]))

        # Random color for the fracture (lighter gray)
        color = random.randint(150, 200)  # Light gray color
        
        # Draw the fracture as a line
        draw.line((start_x, start_y, end_x, end_y), fill=color, width=random.randint(1, 3))
        drawm.line((start_x, start_y, end_x, end_y), fill=250, width=random.randint(1, 3))
    
    # Optionally add some noise for realism
    for _ in range(int(width * height * 0.01)):  # 1% noise
        x = random.randint(0, width - 1)
        y = random.randint(0, height - 1)
        noise_color = random.randint(80, 220)  # Slightly lighter than the background
        image.putpixel((x, y), noise_color)

    # Save the generated image
    image.save('./images/' + save_path)
    mask.save('./masks/'+ save_path)
    print(f"Image saved at {save_path}")

for i in range(2000):
    # Generate and save an example fracture image
    generate_fracture_image(100,100,20,str(i)+'.png')
```
